Cleaning Grant Amount...

```python
colName = "Grant Amount in Rs Crore"
df[colName] = df[colName].str.replace(',','') # removing commas for proper conversion
df[colName] = pd.to_numeric(df[colName], errors='coerce') #forcibly converts "Not Available" to "NaN"
df[colName].replace(np.nan, 0, inplace=True) #replaces nans with 0 for future easy computation
df

```

A1. Getting each state's project costs, then printing max and min... 

```python
df['Location'].astype(str)
locArr = df['Location'].unique()
ser = pd.Series(data=np.zeros(locArr.__len__()),index=locArr)
ser



for i in ser.index:
    ser[i] = df[colName].loc[(df['Location'] == i)].sum() #colName is still as above

print("Max Grant")
print(str(ser.index[ser.argmax()]) + "= "+ str(ser.max())) #Max cost state
print()
#Multiple Min Cost states, therefore..
print("Min grants recieved by the states of")
for i in ser.index:
    if ser[i]==ser.min():
        print(i)

print("who recieved {}".format(ser.min()))
```

A6. Checking Time differences across projects, sorting and loccing 5 highest value rows...

```python
time = (df[colName2]-df[colName1]).dropna()
longest5 = time.sort_values()[-1:-6:-1]
df2 = df.loc[longest5.index,['Project Name', colName1, colName2]]
df2['Difference in Yrs'] = longest5/np.timedelta64(1,'Y')
df2
```

```python
longest = time.sort_values()
df3 = df.loc[longest.index,[colName1,colName2]]
df3['Difference in Yrs'] = longest/np.timedelta64(1,'Y')
df3 = df3[df3['Difference in Yrs'] > 0]
newCol = "Between Yrs"
df3[newCol] =""
df3[newCol].iloc[np.where(df3['Difference in Yrs'] < 7)] = '0-7 years'
df3[newCol].iloc[np.where(df3['Difference in Yrs'] < 6)] = '0-6 years'
df3[newCol].iloc[np.where(df3['Difference in Yrs'] < 5)] = '0-5 years'
df3[newCol].iloc[np.where(df3['Difference in Yrs'] < 4)] = '0-4 years'
df3[newCol].iloc[np.where(df3['Difference in Yrs'] < 3)] = '0-3 years'
df3[newCol].iloc[np.where(df3['Difference in Yrs'] < 2)] = '0-2 years'
df3[newCol].iloc[np.where(df3['Difference in Yrs'] < 1)] = '0-1 years'
df3[newCol].value_counts().sort_values(ascending=True, inplace=True)#.plot(kind='barh', title="Extra time counts of Non-timely Completions", color=['darkred','maroon','firebrick','brown','indianred','lightcoral','red'], xlabel='', ylabel='Number of Projects')
graphFrame2 = df3[newCol].value_counts()
graphFrame2 = graphFrame2.cumsum()



ax = sns.barplot(y=graphFrame2.index, x=graphFrame2, palette='Reds')
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax.set_xlabel("Project Counts")
ax.set_ylabel("Extra Time")
ax.set_title("Extra time counts of Non-timely Completions")
#ax.tick_params(axis='both',labelsize=35)
patches = ax.patches
for i in range(len(patches)):
    y = patches[i].get_y() + patches[i].get_height()/2
    x = patches[i].get_width()+12
    ax.annotate('{:.1f}%'.format(graphFrame2[i]/355*100), (x, y), ha='center')
```

Aggregate Dataframe for testing

```python
as_per_agreement = "Construction Completion Date(As per agreement)"
actual = "Construction Completion Date(Actual)"

df[as_per_agreement] = pd.to_datetime(df[as_per_agreement],errors='coerce')
df[actual] = pd.to_datetime(df[actual],errors='coerce')

data = {'Location': df['Location'], 'Sector': df['Sector'], 'Agreed Completion Date': df[as_per_agreement], 'Actual Completion Date': df[actual]}

completionFrame = pd.DataFrame(data)
completionFrame

a = np.where(df[as_per_agreement]>=df[actual])
b = np.where(df[as_per_agreement]<df[actual])

completionFrame['Completion Status']= "Not Applicable"
completionFrame['Completion Status'].iloc[a] = 'Completed'
completionFrame['Completion Status'].iloc[b] = 'Not Completed'
stateFrame = completionFrame.groupby(by =['Location','Completion Status'])['Completion Status'].count().to_frame().unstack()
stateFrame
completionFrame.to_csv(PATH+"Timely Completion per State.csv", index=False)
```
